```python
import onnx
import onnxoptimizer

input_path = r"C:\VScode\Python\YOLOv\runs-20250525T191005Z-1-001\runs\detect\MekaDis_Yolo12s\weights\last.onnx"
output_path = r"C:\VScode\Python\YOLOv\runs-20250525T191005Z-1-001\runs\detect\MekaDis_Yolo12s\weights\last_optimizedd.onnx"

# Load model
model = onnx.load(input_path)

# Lihat pass yang tersedia (opsional)
print("Passes yang tersedia:")
print(onnxoptimizer.get_available_passes())

# Pilih pass yang aman
passes = [
    "eliminate_identity",
    "eliminate_nop_dropout",
    "eliminate_nop_monotone_argmax",
    "eliminate_nop_pad",
    "eliminate_unused_initializer",
    "extract_constant_to_initializer",
    "fuse_add_bias_into_conv",
]

# Optimasi
optimized_model = onnxoptimizer.optimize(model, passes)

# Simpan
onnx.save(optimized_model, output_path)
print(f"✅ Model berhasil dioptimalkan dan disimpan di {output_path}")
```
